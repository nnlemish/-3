```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
possible_encodings = ['utf-8', 'latin-1', 'cp1251']

for encoding in possible_encodings:
    try:
        # Спроба зчитати файл з поточним кодуванням
        data = pd.read_csv(r'C:\Users\admin\Desktop\Настя\spotify-2023.csv', encoding=encoding)
        print("Файл успішно зчитано з кодуванням:", encoding)
        # Якщо вдалося зчитати, вийти з циклу
        break
    except UnicodeDecodeError:
        print("Не вдалося зчитати з кодуванням:", encoding)
```

```python
#Завдання полягає в тому, щоб побудувати горизонтальну стовпчикову діаграму, на якій відображаються 10 найпопулярніших пісень на Spotify у 2023 році за кількістю прослуховувань. 

top_10 = data[data.released_year == 2023].sort_values('streams', ascending = False).head(10)

plt.barh(top_10['track_name'], top_10['streams'], color='skyblue')
plt.xlabel('Streams', fontsize = 25)
plt.ylabel('Track name', fontsize = 25)
plt.title('10 найпопулярніших пісень у 2023 році за кількістю прослуховувань', fontsize = 35)

plt.show()
```

```python
#Рейтинг артистів на Spotify: Створіть стовпчикову діаграму, на якій відображаються 10 найпопулярніших артистів на Spotify у 2023 році за кількістю слухачів. 

name10 = data[data.released_year == 2023].sort_values('streams', ascending = False).head(10)

plt.bar(name10['artist(s)_name'], name10['streams'])
plt.xlabel('Артисти', fontsize = 20)
plt.ylabel('Кількість прослуховувань', fontsize = 20)
plt.title('10 найпопулярніших артистів на Spotify у 2023 році за кількістю слухачів', fontsize = 30)
plt.xticks(rotation = 90)

plt.show()
```
